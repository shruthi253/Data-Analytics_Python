```python
# Load only first 1000000 columns to avoid memory error
df = ords_prods_merge[:1000000]
```

```python
# Define function to label product proce range
def price_label(row):
  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df.head()
```

```python
# apply function to create new column
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Define conditions based on frequency table above
result = []

for value in ords_prods_merge['orders_day_of_week']:
    if value == 0.0:
        result.append("Busiest day")
    elif value == 4.0:
        result.append("Least busy day")
    else:
        result.append("Regularly busy day")
```

```python
# Find 2 busiest days and 2 slowest days of the week
result = []

for value in ords_prods_merge['orders_day_of_week']:
    if value in (0, 1):
        result.append("Busiest days")
    elif value in (3,4):
        result.append("Least busy days")
    else:
        result.append("Regularly busy days")
```

    #step 4: Busiest hour of the day:Most orders, Average orders, and Fewest orders. # column: busiest_period_of_day

```python
ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Setting busiest hours as period with orders above 100000 and above 100000
result = []

for value in ords_prods_merge['order_hour_of_day']:
    if value in (10 , 11 , 14 , 15 , 13 ,12 ,16 , 9.0, 17 , 8.0 , 18.0 , 19.0 ):
        result.append("Most orders") # orders above 1000000
    elif value in (20.0, 7.0 , 21.0 , 22.0 , 23.0 ,6.0 ,0.0 , 1.0 ):
        result.append("Average orders") # orders above 100000
    else:
        result.append("Fewest orders")
```

```python
# adding new column for busiest period of the day labels
ords_prods_merge['busiest_period_of_day'] = result
```

```python
# Check frequency : The sum of each group seems equal 
ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```

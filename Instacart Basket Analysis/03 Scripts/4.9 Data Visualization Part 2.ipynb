#scatterplot between age and income

```python
# working with the samples of merged df to avoid memory issues
merged_4 = small[['age', 'income']]
```

```python
# create scatterplot
scatterplot_age_income = sns.scatterplot(x = 'age', y = 'income', data = merged_4)
```

#it is evident from the plot that most people have the income 200000 and people of age >=40 have income above 400000, which makes them have more power of expenditure compared to people under 40. Group 20-40 also has some outliers with income from 200000 to 400000 which give them more power of expenditure compared to other people of their own age group. Overall it can be said that as age/experience increase there is chance of high income and hence power of expenditure.

```python
#exporting scatterplot_age_income
scatterplot_age_income.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'scatterplot_age_income.png'))
```

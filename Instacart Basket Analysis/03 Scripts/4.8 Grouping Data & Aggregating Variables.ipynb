```python
# create max_order column and use it to create loyalty flag

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Setting loyalty flag

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40 , 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10 , 'loyalty_flag'] = 'New customer'
```

    # Step 6: Spending flag based on mean prices spent

```python
# Aggregated mean for entire dataset for user_id
ords_prods_merge.groupby('user_id').agg({'prices': ['mean']})
```

```python
ords_prods_merge['mean_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge['mean_price'].value_counts(dropna = False)
```

```python
# Flag based on mean price spend
ords_prods_merge.loc[(ords_prods_merge['mean_price'] < 10), 'spend_flag'] = 'Low Spender'
```

```python
ords_prods_merge.loc[(ords_prods_merge['mean_price'] >= 10), 'spend_flag'] = 'High Spender'
```

```python
ords_prods_merge[['user_id', 'loyalty_flag', 'mean_price', 'spend_flag']].head(20)
```

```python
ords_prods_merge['spend_flag'].value_counts(dropna = False)
```

    # Step 7: Labelling based on customer behaviour: days before previous order

```python
# calculate median
ords_prods_merge['median_days_prev_order'] = ords_prods_merge.groupby(['user_id'])['days_since_previous_order'].transform(np.median)
```

```python
# Setting label:
ords_prods_merge.loc[ords_prods_merge['median_days_prev_order'] > 20 , 'regularity_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_prev_order'] > 10) & (ords_prods_merge['median_days_prev_order'] <= 20) , 'regularity_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_prev_order'] <= 10 , 'regularity_flag'] = 'Frequent customer'
```

```python
ords_prods_merge['regularity_flag'].value_counts(dropna = False)
```
